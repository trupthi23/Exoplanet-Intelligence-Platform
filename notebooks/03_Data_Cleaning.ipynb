### Phase 3B – Missing Value Strategy

```python
missing = pd.DataFrame({
    "Missing Count": df.isnull().sum(),
    "Missing %": (df.isnull().sum() / len(df)) * 100
})

missing.sort_values("Missing %", ascending=False)
```

```python
essential_columns = [
    "planet_name",
    "host_star",
    "discovery_method",
    "discovery_year"
]

important_columns = [
    "planet_radius_earth",
    "orbital_period_days",
    "star_temperature",
    "star_radius",
    "star_mass",
    "distance_pc"
]

optional_columns = [
    "planet_mass_earth",
    "equilibrium_temperature",
    "right_ascension",
    "declination"
]
```

```python
print(df[essential_columns].isnull().sum())
```

```python
print(df[important_columns].isnull().sum())
```

```python
print(df[optional_columns].isnull().sum())
```

### Missing Value Strategy

- Essential columns: Must not contain missing values.
- Important columns: Keep missing values for now; they may be filtered for specific analyses or models.
- Optional columns: Retain missing values because they represent measurements that have not yet been observed, not data entry errors.
